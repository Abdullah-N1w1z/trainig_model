```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import os
```

```python
DATA_DIR = "data/raw/ohlcv"
os.makedirs(DATA_DIR, exist_ok=True)
```

```python
tickers = [
    "AAPL", "MSFT", "GOOGL", "AMZN", "TSLA",    # Tech
    "JPM", "BAC", "WFC",                        # Financials
    "XOM", "CVX",                               # Energy
    "JNJ", "PFE",                               # Healthcare
    "KO", "PEP",                                # Consumer Staples
    "MCD", "SBUX",                              # Discretionary
    "BA", "CAT",                                # Industrials
    "CSCO", "IBM",                              # Tech/IT
    "V", "MA",                                  # Fintech
    "T", "VZ",                                  # Telecom
    "WMT", "COST",                              # Retail
    "DIS", "NFLX",                              # Media
    "GILD", "AMGN",                             # Biotech
    "RTX", "LOW", "GE"                          # Misc
]

START_DATE = "2015-06-01"
END_DATE = "2025-06-01"


for ticker in tickers:
    print(f" Downloading {ticker}...")
    try:
        df = yf.download(ticker, start=START_DATE, end=END_DATE, progress=False)
        if df.empty:
            print(f"No data for {ticker}")
        else:
            if isinstance(df.columns, pd.MultiIndex):
                df.columns = df.columns.get_level_values(0)
            df.reset_index(inplace=True)
            df = df[["Date", "Open", "High", "Low", "Close", "Volume"]]
            df = df.dropna() 
            df[["Open", "High", "Low", "Close"]] = df[["Open", "High", "Low", "Close"]].astype(float)
            df["Volume"] = df["Volume"].astype(int)
            df.to_csv(os.path.join(DATA_DIR, f"{ticker}.csv"), index=False)
            print(f"Saved: {ticker}.csv")
    except Exception as e:
        print(f"Error downloading {ticker}: {e}")
```

```python
import os
import pandas as pd
import numpy as np

from ta.trend import SMAIndicator, EMAIndicator, MACD, ADXIndicator, CCIIndicator
from ta.momentum import RSIIndicator, StochasticOscillator, ROCIndicator
from ta.volatility import BollingerBands, AverageTrueRange
from ta.volume import OnBalanceVolumeIndicator

DATA_DIR = "data/raw/ohlcv"
OUT_DIR = "stock_features"
os.makedirs(OUT_DIR, exist_ok=True)

def rolling_slope(series, window):
    slopes = []
    for i in range(len(series)):
        if i < window - 1:
            slopes.append(np.nan)
        else:
            y = series[i-window+1:i+1]
            x = np.arange(window)
            A = np.vstack([x, np.ones(window)]).T
            m, _ = np.linalg.lstsq(A, y, rcond=None)[0]
            slopes.append(m)
    return pd.Series(slopes, index=series.index)

def calc_features(df):
    df = df.copy()
    # Returns
    df["return_1d"] = df["Close"].pct_change(1)
    df["return_5d"] = df["Close"].pct_change(5)
    df["return_10d"] = df["Close"].pct_change(10)
    df["log_return"] = np.log(df["Close"] / df["Close"].shift(1))

    # SMA & EMA
    df["sma_10"] = SMAIndicator(close=df["Close"], window=10).sma_indicator()
    df["sma_20"] = SMAIndicator(close=df["Close"], window=20).sma_indicator()
    df["sma_50"] = SMAIndicator(close=df["Close"], window=50).sma_indicator()
    df["sma_100"] = SMAIndicator(close=df["Close"], window=100).sma_indicator() 
    df["sma_150"] = SMAIndicator(close=df["Close"], window=150).sma_indicator()
    df["sma_200"] = SMAIndicator(close=df["Close"], window=200).sma_indicator()
    df["ema_9"] = EMAIndicator(close=df["Close"], window=9).ema_indicator()
    df["ema_12"] = EMAIndicator(close=df["Close"], window=12).ema_indicator()
    df["ema_20"] = EMAIndicator(close=df["Close"], window=20).ema_indicator()
    df["ema_26"] = EMAIndicator(close=df["Close"], window=26).ema_indicator()
    df["ema_50"] = EMAIndicator(close=df["Close"], window=50).ema_indicator()

    # MACD
    macd_9 = MACD(close=df["Close"], window_slow=26, window_fast=12, window_sign=9)
    df["macd_9"] = macd_9.macd()
    df["macd_signal_9"] = macd_9.macd_signal()
    df["macd_hist_9"] = macd_9.macd_diff()
    macd_20 = MACD(close=df["Close"], window_slow=26, window_fast=12, window_sign=20)
    df["macd_20"] = macd_20.macd()
    df["macd_signal_20"] = macd_20.macd_signal()
    df["macd_hist_20"] = macd_20.macd_diff()

    # ROC
    df["roc_10"] = ROCIndicator(close=df["Close"], window=10).roc()

    # Slope
    df["slope_10"] = rolling_slope(df["Close"], 10)

    # Volatility
    df["rolling_std_10"] = df["Close"].rolling(10).std()
    df["rolling_std_20"] = df["Close"].rolling(20).std()
    df["rolling_std_50"] = df["Close"].rolling(50).std()
    df["rolling_std_200"] = df["Close"].rolling(200).std()
    df["atr_14"] = AverageTrueRange(high=df["High"], low=df["Low"], close=df["Close"], window=14).average_true_range()

    # RSI
    df["RSI_9"] = RSIIndicator(close=df["Close"], window=9).rsi()
    df["RSI_14"] = RSIIndicator(close=df["Close"], window=14).rsi()
    df["RSI_30"] = RSIIndicator(close=df["Close"], window=30).rsi()

    # Bollinger Bands
    bb_20 = BollingerBands(close=df["Close"], window=20)
    df["bollinger_high"] = bb_20.bollinger_hband()
    df["bollinger_low"] = bb_20.bollinger_lband()
    df["bb_width"] = df["bollinger_high"] - df["bollinger_low"]

    bb_50 = BollingerBands(close=df["Close"], window=50)
    df["BB_High_50"] = bb_50.bollinger_hband()
    df["BB_Low_50"] = bb_50.bollinger_lband()

    # Z-score
    df["zscore_20"] = (df["Close"] - df["Close"].rolling(20).mean()) / (df["Close"].rolling(20).std() + 1e-6)

    # Volume
    df["volume_change_1d"] = df["Volume"].pct_change(1)
    df["volume_avg_20"] = df["Volume"].rolling(20).mean()
    df["volume_spike"] = df["Volume"] / (df["volume_avg_20"] + 1e-6)
    obv = OnBalanceVolumeIndicator(close=df["Close"], volume=df["Volume"]).on_balance_volume()
    df["obv"] = obv
    df["OBV_20"] = obv.rolling(window=20).mean()
    df["OBV_50"] = obv.rolling(window=50).mean()
    df["OBV_100"] = obv.rolling(window=100).mean()

    # Candlestick
    df["candle_body"] = df["Close"] - df["Open"]
    df["upper_wick"] = df["High"] - df[["Close", "Open"]].max(axis=1)
    df["lower_wick"] = df[["Close", "Open"]].min(axis=1) - df["Low"]
    df["candle_range"] = df["High"] - df["Low"]
    df["body_ratio"] = df["candle_body"] / (df["candle_range"] + 1e-6)
    df["doji_flag"] = (df["candle_body"].abs() / (df["candle_range"] + 1e-6)) < 0.1

    # Lag Features
    for i in range(1, 6):
        df[f"close_lag{i}"] = df["Close"].shift(i)
        df[f"return_lag{i}"] = df["return_1d"].shift(i)

    # Stochastic Oscillator (added for windows 9, 14, 20)
    stoch_9 = StochasticOscillator(high=df['High'], low=df['Low'], close=df['Close'], window=9, smooth_window=3)
    df['Stoch_Osc_9'] = stoch_9.stoch()
    df['Stoch_Signal_9'] = stoch_9.stoch_signal()

    stoch_14 = StochasticOscillator(high=df['High'], low=df['Low'], close=df['Close'], window=14, smooth_window=3)
    df['Stoch_Osc_14'] = stoch_14.stoch()
    df['Stoch_Signal_14'] = stoch_14.stoch_signal()

    stoch_20 = StochasticOscillator(high=df['High'], low=df['Low'], close=df['Close'], window=20, smooth_window=3)
    df['Stoch_Osc_20'] = stoch_20.stoch()
    df['Stoch_Signal_20'] = stoch_20.stoch_signal()

    # Fibonacci Retracement Levels (example for 50-day window)
    window = 50
    high = df['Close'].rolling(window).max()
    low = df['Close'].rolling(window).min()
    df['fib_236'] = high - 0.236 * (high - low)
    df['fib_382'] = high - 0.382 * (high - low)
    df['fib_500'] = high - 0.500 * (high - low)
    df['fib_618'] = high - 0.618 * (high - low)
    df['fib_786'] = high - 0.786 * (high - low)

    return df

for fname in os.listdir(DATA_DIR):
    if fname.endswith('.csv'):
        df = pd.read_csv(os.path.join(DATA_DIR, fname))
        features_df = calc_features(df)
        features_df.to_csv(os.path.join(OUT_DIR, fname), index=False)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the features DataFrame (replace 'AAPL.csv' with your file if needed)
df = pd.read_csv('stock_features/AAPL.csv')

# Select only feature columns (exclude Date, Open, High, Low, Close, Volume if present)
feature_cols = [col for col in df.columns if col not in ['Date', 'Open', 'High', 'Low', 'Close', 'Volume']]
features_df = df[feature_cols]

# Compute correlation matrix
corr = features_df.corr()

# Plot heatmap with 1 decimal place and larger size
plt.figure(figsize=(24, 20))
sns.heatmap(corr, cmap='coolwarm', center=0, annot=True, fmt=".1f")
plt.title('Feature Correlation Heatmap', fontsize=22)
plt.tight_layout()
plt.savefig('feature_heatmap.png', dpi=300)
plt.show()
```
